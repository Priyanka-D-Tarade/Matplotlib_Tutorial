```python
import matplotlib.pyplot as plt     #import library
%matplotlib inline                 
#prints the graph
```

# Piechart====>requires categorical as well as numerical data

```python
areas=["Marketing","sales","Development","HR","Customer support"]
budget=[2.5,6,35,6,4]   #2 types of data required(categorical and numerical)
```

```python
plt.pie(budget,labels=areas)
plt.show()
```

```python
plt.pie(budget,labels=areas,autopct="%0.1f%%")
plt.show()
```

# Scatterplot==>reqired both numerical data

```python
a=[23,14,15,17,18,19,23,18,16,20]
b=[4,5,7,8,9,7,8,9,2,1]
plt.scatter(a,b)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs salary graph")
```

# Lineplot====>requires Both numerical data.

```python
product_id=[1,2,3,4,5,6,7,8,9]
online_price=[33,67,43,65,80,90,32,32,43]
offline_price=[30,60,30,40,10,40,60,80,90]
chor_bazar=[60,70,20,20,10,20,30,40,50]
```

```python
plt.plot(product_id,online_price, color="blue", marker="D",markersize=10,label="online")
plt.plot(product_id,offline_price, color="red", marker="*",markersize=10,label="offline")
plt.plot(product_id,chor_bazar, color="green", marker="+",markersize=10, label="chori")

plt.xlabel("product")
plt.ylabel("price comparison")
plt.show()
plt.legend()
```

# Barchart===>requires both numerical and discreate types of data

```python
yr=[2012,2013,2014,2015,2016,2017,2018]
price=[45000,50000,35000,60000,40000,70000,80000]
plt.bar(yr,price,color="Blue")
plt.xlabel("year")
plt.ylabel("prices in yrs")
plt.title("bike car comparison")
```
